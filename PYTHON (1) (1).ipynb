```python
# BMI Calculator
# calculates the Body Mass Index of an individual
# Formula: BMI = Weight(kg)/Height(m)**2
```

```python
# convert input to a float data type
weight = float(input('Weight (kg): '))
height = float(input('Height (m): '))

# Calculate BMI
bmi = round(weight / (height**2), 2)

# Using the IF statements to categorise the BMI
if bmi < 18:
    print(f'Your Body Mass Index is {bmi}. You are Underweight!')
elif bmi <= 25:
    print(f'Your Body Mass Index is {bmi}. You have a Normal Weight!')
else:
    print(f'Your Body Mass Index is {bmi}. You are Overweight OR Obese!')

# BMI Chart Message
msg = '''
Body Mass Index Chart:
Between 18 - 25 = Normal Weight
Less than (<) 18 = Underweight
Greater than (>) 25 = Overweight OR Obese
'''

print(msg)
```

```python
# Civil Servant Retirement Predictor
```

```python
# Get the civil servant's details
title = input('Title: ')
first_name = input('First Name: ')
last_name = input('Last Name: ')

# Convert input to an integer data type
year_commenced = int(input('Year Commenced Civil Service: '))
birth_year = int(input('Birth Year: '))
current_year = int(input('Current Year: '))

# Calculate the civil servant's current age
current_age = current_year - birth_year

# Calculate years of active service
years_of_service = current_year - year_commenced

# Set the official retirement age
retirement_age = 60

# Calculate the retirement year
retirement_year = birth_year + retirement_age

# Calculate years left before retirement
years_left = retirement_year - current_year

print()
print(f'Hello {title} {first_name} {last_name}.')
print(f'You are {current_age} years old and you will be retiring in the year {retirement_year} at the age of {retirement_age}.')
print(f'You have spent {years_of_service} years of active service.')
print(f'You have {years_left} more years of active service left.')
print('Thank you for your service.')
```

```python
# For Loop
# Calculate the Total Price in the List
```

```python
# Create a list to store all the prices
price = [25, 257, 212, 116, 99, 35]

# Create a variable to store the total price
# We start from 0 because we haven't added anything yet
total = 0

# Use a for loop to go through each price in the list
for amount in price:

    # Add each price to the running total
    total = total + amount

# Display the final total price
print(f'The total price is {total}')
```

```python
# Dictionary Method
# For Loop
# Create a Dictionary of all the African countries
```

```python
# Create a dictionary to store African countries,
# their capitals, GDP, and region
africa = {
    'Nigeria': {
        'Capital': 'Abuja',
        'GDP': '363 Billion USD',
        'Region': 'West Africa'
    },

    'Kenya': {
        'Capital': 'Nairobi',
        'GDP': '113 Billion USD',
        'Region': 'East Africa'
    },

    'Egypt': {
        'Capital': 'Cairo',
        'GDP': '396 Billion USD',
        'Region': 'North Africa'
    },

    
    'South Africa': {
        'Capital': 'Pretoria',
        'GDP': '380 Billion USD',
        'Region': 'Southern Africa'
    },

    'Ghana': {
        'Capital': 'Accra',
        'GDP': '76 Billion USD',
        'Region': 'West Africa'
    }
}

# Use a for loop to go through each country
# and its information in the dictionary
for country, details in africa.items():

    # Display the country's name
    print(f'Country: {country}')

    # Display the country's capital
    print(f'Capital: {details['Capital']}')

    # Display the country's GDP
    print(f'GDP: {details['GDP']}')

    # Display the country's region
    print(f'Region: {details['Region']}')

    # Print a blank line before displaying the next country
    print()
```
